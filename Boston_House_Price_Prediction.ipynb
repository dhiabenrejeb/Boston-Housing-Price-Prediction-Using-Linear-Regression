## Features Scaling

### Standard Scaler

```python
X=H[['RM']].values
y=H['MEDV'].values
```

```python
from sklearn.preprocessing import StandardScaler
x_std=StandardScaler().fit_transform(X)
y_std=StandardScaler().fit_transform(y[:,np.newaxis]).flatten()
```

### Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler

x_minmax=MinMaxScaler().fit_transform(X)
y_minmax=MinMaxScaler().fit_transform(y[:,np.newaxis]).flatten()
```

```python
class LinearRegressionGD(object):
    def __init__(self, eta=0.1, n_iter=10):
        self.eta = eta
        self.n_iter = n_iter
    def fit(self,X,y):
        self.w_ = np.zeros(1+X.shape[1])
        self.errors_ = []
         
        for i in range(self.n_iter):
            output = X.dot(self.w_[1:])+self.w_[0]
            errors = y - output
            self.w_[1:] += (self.eta) * np.dot(X.T, errors)
            self.w_[0] += (self.eta) * errors.sum()
            e = (errors**2).sum() / 2.0
            self.errors_.append(e)
        return self
    def predict(self, X):
        return X.dot(self.w_[1:])+self.w_[0]
        
```

```python
hp=LinearRegressionGD(eta=0.001, n_iter=50)
hp.fit(x_std,y_std)
```

```python
plt.figure(1)
plt.scatter(x_std,y_std)
p=hp.predict(x_std)
plt.plot(x_std,p,color='red')
```

```python
print('weight using standard scaling is',hp.w_)
```

```python
print(hp.errors_)
```

```python
plt.plot( range(1, len(hp.errors_)+1),hp.errors_)
plt.xlabel('Epochs')
plt.ylabel('errors')
plt.show()
```

```python
ss=LinearRegressionGD(eta=0.001, n_iter=50)
ss.fit(x_minmax,y_minmax)
plt.figure(2)
plt.scatter(x_minmax,y_minmax)
mn=ss.predict(x_minmax)
plt.plot(x_minmax,mn,color='red')
```

```python
plt.plot( range(1, len(ss.errors_)+1),hp.errors_)
plt.xlabel('Epochs')
plt.ylabel('errors')
plt.show()
```

```python
print('weight using min max scaling is',ss.w_)
```
